```python
# Import dependencies
from path import Path
import pandas as pd
import numpy as np
```

```python
# Load csv into df
data = Path('../Raw_Resources/wine_main.csv')
df = pd.read_csv(data)
df.head()
```

```python
# Count # of rows - raw
index = df.index
no_rows = len(index)
print(no_rows)
```

```python
#  Drop the NaNs
df.dropna(inplace=True)
df.head(10)
```

```python
# Count # of rows - cleaned
index = df.index
no_rows = len(index)
print(no_rows)
```

```python
df = df.sample(n=1000)
```
